```python
parameters = [[7], [3, 8], [8, 1, 0], [2, 7, 4, 4], [4, 5, 2, 6, 5]]
#answer = 30
```

# My solution

```python
def my_solution(triangle):
    
    j = 1
    while j < len(triangle):
        upper_line = triangle[-j-1]
        bottom_line = triangle[-j]

        for i in range(len(bottom_line)-1):
            upper_line[i] += max(bottom_line[i],bottom_line[i+1])

        bottom_line = upper_line
        j += 1

    return bottom_line[0]
```

# Other solution

```python
solution = lambda t, l = []: max(l) if not t else solution(t[1:], [max(x,y)+z for x,y,z in zip([0]+l, l+[0], t[0])])
```

```python
import time 

print("my solution")
start = time.time()
print( 'answer = ' ,my_solution(parameters), 'runtime = ' , time.time() - start )


print("best solution 1")
start = time.time()
print( 'answer = ' ,solution(parameters), 'runtime = ' , time.time() - start )
```
