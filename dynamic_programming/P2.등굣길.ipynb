```python
m = 4 
n = 3
puddles = [[2, 2]]
# answer  4 
```

```python
import numpy as np

map_ = np.zeros([n,m],dtype=np.int16)
map_[0,:] = 1
map_[:,0] = 1

for puddle in puddles:
    map_[puddle[0]-1][puddle[1]-1] = -1


map_


for a in range(1,n):
    for b in range(1,m):
        if map_[a,b] == -1:
            continue
        if map_[a-1,b] == 1:
            map_[a,b] += 1
        if map_[a,b-1] == 1:
            map_[a,b] += 1

print(map_)

answer = 0
if map_[n-2,m-1] != -1:
    answer += map_[n-2,m-1]
if map_[n-1,m-2] != -1:
    answer += map_[n-1,m-2]    
#answer = int(map_[n-2,m-1] + map_[n-1,m-2])
print(answer)
```
