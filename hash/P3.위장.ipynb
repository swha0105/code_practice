```python
parameters = [] 
parameters.append([["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]]) 
parameters.append([["crow_mask", "face"], ["blue_sunglasses", "face"], ["smoky_makeup", "face"]])
#answer 5,3
```

# my solution

```python
def my_solution(clothes):
    import collections 
    import itertools
    
    clothes_dict = collections.defaultdict(list)
    number_list = []
    for value,key in clothes:
        clothes_dict[str(key)].append(value)

    answer =  1

    for key in clothes_dict:
        tmp_answer =  len(list(itertools.combinations(clothes_dict[str(key)],1))) + 1
        answer = (answer*tmp_answer)

    # answer =  1

    # for key in clothes_dict:
    #     answer *=  len(list(itertools.combinations(clothes_dict[str(key)],1))) + 1
        

    answer = answer - 1
    return answer
```

# Best solution

```python
def solution(clothes):
    from collections import Counter
    from functools import reduce
    cnt = Counter([kind for name, kind in clothes])
    answer = reduce(lambda x, y: x*(y+1), cnt.values(), 1) - 1
    return answer
```

```python
import time 

print("my solution")
for input_ in parameters:
    start = time.time()
    print( 'answer = ' ,my_solution(input_), 'runtime = ' , time.time() - start )


print("best solution")
for input_ in parameters:
    start = time.time()
    print( 'answer = ' ,solution(input_), 'runtime = ' , time.time() - start )

```
