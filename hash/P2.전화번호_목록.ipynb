# 문제 핵심
### list에 있는 string들 끼리 접두어가 되면 안됨. 

```python
parameters = [] 
parameters.append(["119", "97674223", "1195524421"]) 
parameters.append(["123", "456", "789"])
parameters.append(["12", "123", "1235", "567", "88"])

# False, True, False
```

# My code

```python
def my_solution(phone_book):
    phone_book = sorted(phone_book,key=len)

    for _ in phone_book:
        ref = phone_book.pop(0)
        
        phone_number_list = [s for s in phone_book if ref in s[:len(ref)]]
        
        if len(phone_number_list) >= 1:
            answer = False
            break
    else:
        answer = True
    return answer
```

# best solution 1

```python
def solution1(phoneBook):
    phoneBook = sorted(phoneBook)

    for p1, p2 in zip(phoneBook, phoneBook[1:]):
        if p2.startswith(p1):
            return False
    return True
```

# best solution 2

```python
def solution2(phone_book):
    answer = True
    hash_map = {}
    for phone_number in phone_book:
        hash_map[phone_number] = 1
    for phone_number in phone_book:
        temp = ""
        for number in phone_number:
            temp += number
            if temp in hash_map and temp != phone_number:
                answer = False
    return answer    
```

```python
import time 

print("my solution")
for input_ in parameters:
    start = time.time()
    print( 'answer = ' ,my_solution(input_), 'runtime = ' , time.time() - start )


print("best solution 1")
for input_ in parameters:
    start = time.time()
    print( 'answer = ' ,solution1(input_), 'runtime = ' , time.time() - start )

print("best solution 2")
for input_ in parameters:
    start = time.time()
    print( 'answer = ' ,solution2(input_), 'runtime = ' , time.time() - start )

```

### Solution1은 string type에 있는 startswith 라는 내장 함수 사용. 당연히 endswith도 있다.
### 이 함수는 p2.startswith는 p2의 string 을 받아 순차적으로 읽어서 p1이 종료 될때까지 모두 일치하면 True를 반환한다. 
### 내 코드와 개념상 일치하지만 내장함수 검색해보는걸 잊지말자

### solution2 는 hash를 억지로 쓴느낌인거 같다.
### 내가 잘몰라서 그런지 굳이 hash를 안써도 돌아갈꺼같은데.. 나중에 다시봐야겠다
