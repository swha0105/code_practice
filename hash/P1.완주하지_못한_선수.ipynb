# 문제 핵심 

### 리스트 안, 어떤 요소가 몇개가 있는지 판단할때는
### Loop iteration이 아닌, 파이썬 내장함수 collections을 사용한다.

```python
parameters = [] 
parameters.append(["leo", "kiki", "eden"]) 
parameters.append(["marina", "josipa", "nikola", "vinko", "filipa"])
parameters.append(["mislav", "stanko", "mislav", "ana"])

parameters_2 = [] 
parameters_2.append(["eden", "kiki"])
parameters_2.append(["josipa", "filipa", "marina", "nikola"])
parameters_2.append(["stanko", "ana", "mislav"])

#answer "leo","vinko","mislav"
```

# My code

```python
def my_solution(participant, completion):
    import collections
    
    p = collections.Counter(participant)
    c = collections.Counter(completion)

    p.subtract(c)
    answer = []
    for name in p:
        if p[str(name)] >=1 :
            answer.append(str(name)) 
     
    answer = ','.join(answer)
    return answer
```

# Best solution 

```python
def solution(participant, completion):
    answer = collections.Counter(participant) - collections.Counter(completion)
    return list(answer.keys())[0]
```

```python
import time 

print("my solution")
for input_,input_2 in zip(parameters,parameters_2):
    start = time.time()
    print( 'answer = ' ,my_solution(input_,input_2), 'runtime = ' , time.time() - start )


print("best solution")
for input_,input_2 in zip(parameters,parameters_2):
    start = time.time()
    print( 'answer = ' ,my_solution(input_,input_2), 'runtime = ' , time.time() - start )
```
